```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Preprocessing
df['statement'] = df['statement'].str.lower()  # Convert statements to lowercase

# Feature Extraction
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['statement'])  # Extract TF-IDF features from the statements

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, df['status'], test_size=0.2, random_state=42)

# Model Training
model = LogisticRegression()
model.fit(X_train, y_train)

# Model Evaluation
accuracy = model.score(X_test, y_test)
print("Accuracy:", accuracy)

# Prediction
new_statements = ["I feel anxious", "I am happy today"]
new_statements_preprocessed = [statement.lower() for statement in new_statements]
new_statements_features = vectorizer.transform(new_statements_preprocessed)
predictions = model.predict(new_statements_features)
print("Predictions:", predictions)
```

-Precision: Measures the proportion of correctly predicted instances for each class. 
Higher precision values indicate fewer false positives. The highest precision is observed for "Personality disorder" (0.96), "Normal" (0.90), and "Bipolar" (0.92).

.Recall: Measures the proportion of actual instances of each class that were correctly predicted. 
Higher recall values indicate fewer false negatives. The highest recall is observed for "Normal" (0.97), "Anxiety" (0.85), and "Depression" (0.81).

- F1-score: The harmonic mean of precision and recall, providing a balanced measure of model performance. 
The highest F1-scores are observed for "Normal" (0.93), "Anxiety" (0.87), and "Bipolar" (0.84).

- Support: Represents the number of instances in each class.

- Accuracy: The overall accuracy of the model is 0.84, indicating the proportion of correctly classified instances out of the total number of instances.

- Macro average F1-score: 0.81, indicating the overall performance of the model across all classes.

- Weighted average F1-score: 0.83, considering the class imbalance.
